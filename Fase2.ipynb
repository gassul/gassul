```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos=pd.read_csv('DatosA2.csv', header=2,index_col=0)
df_datos=pd.DataFrame(datos)
df_datos
```

```python
serieD = df_datos ['BMI']
grafico = serieD
fig, ax = plt.subplots()
ax.plot(grafico, linewidth=7)

ax.set_title("Monitoreo de salud", fontsize=14)
ax.set_xlabel("Octubre", fontsize=14)
ax.set_ylabel("BMI", fontsize=14)

ax.tick_params(axis='both', labelsize=2)


plt.show()
```
